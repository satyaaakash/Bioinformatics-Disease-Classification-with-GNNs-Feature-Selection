```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
def perform_pca_analysis(file_path, target_column, n_components=10):
    """
    Perform PCA analysis on the dataset.
    - file_path: Path to the dataset.
    - target_column: Name of the target column.
    - n_components: Number of principal components to compute.
    """
    # Load the dataset
    df = pd.read_csv(file_path)
    
    # Separate features and target
    X = df.drop(columns=[df.columns[0], target_column])  # Exclude sample IDs and target column
    y = df[target_column]
    
    # Handle non-numeric columns by encoding them
    if X.select_dtypes(include=["object"]).shape[1] > 0:
        X = X.apply(pd.to_numeric, errors='coerce')

    # Handle missing values
    X = X.fillna(X.mean())
    
    # Standardize the features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Perform PCA
    pca = PCA(n_components=n_components)
    principal_components = pca.fit_transform(X_scaled)
    
    # Variance explained by each component
    explained_variance = pca.explained_variance_ratio_
    cumulative_variance = np.cumsum(explained_variance)
    
    # Plot variance explained
    plt.figure(figsize=(6, 4))
    plt.bar(range(1, n_components + 1), explained_variance, alpha=0.6, label='Individual Explained Variance')
    plt.step(range(1, n_components + 1), cumulative_variance, where='mid', label='Cumulative Explained Variance')
    plt.title('PCA Variance Explained')
    plt.xlabel('Principal Components')
    plt.ylabel('Variance Explained')
    plt.legend()
    plt.show()
    
    # Scatter plot of the first two principal components
    plt.figure(figsize=(6, 4))
    plt.scatter(principal_components[:, 0], principal_components[:, 1], c=pd.factorize(y)[0], cmap='viridis', alpha=0.7)
    plt.colorbar(label='Classes')
    plt.title('Scatter Plot of First Two Principal Components')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.show()
    
    # Return PCA results
    return {
        "Principal Components": principal_components,
        "Explained Variance": explained_variance,
        "Cumulative Variance": cumulative_variance
    }
```

```python

# Example usage
if __name__ == "__main__":
    # File path to the dataset
    dataset_file = "preprocessed/preprocessed_with_target_GSE4290.csv"  # Replace with your dataset
    target_column = "Target"  # Replace with the actual target column name

    print(f"Performing PCA analysis on {dataset_file}...")
    pca_results = perform_pca_analysis(dataset_file, target_column, n_components=10)

    # Display explained variance
    print("Explained Variance by Principal Components:", pca_results["Explained Variance"])
    print("Cumulative Variance by Principal Components:", pca_results["Cumulative Variance"])
```
